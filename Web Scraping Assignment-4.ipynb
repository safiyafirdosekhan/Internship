<b>1. Scrape the details of most viewed videos on YouTube from Wikipedia.    
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos      
You need to find following details:       
A) Rank     
B) Name       
C) Artist      
D) Upload date      
E) Views</b>

```python
import selenium
import pandas as pd
from selenium import webdriver

import warnings
warnings.filterwarnings('ignore')

from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
# Let's first connect to the driver

driver = webdriver.Chrome(r"C:\Users\safiy\Downloads\chromedriver_win32 (2)\chromedriver.exe")
driver.maximize_window()
time.sleep(2)
```

```python
# Opening Youtube on automated chrome browser

driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
# Creating empty lists

Rank = []
Name = []
Artist = []
Upload_date = []
Views = []
```

```python
# Scraping Rank

Rank_tag = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")

for i in Rank_tag:
    Rank.append(i.text)
```

```python
# Scraping Name

Name_tag = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")

for i in Name_tag:
    Name.append(i.text.split('[')[0])
```

```python
# Scraping Artist

Artist_tag = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")

for i in Artist_tag:
    Artist.append(i.text)
```

```python
# Scraping Upload date

Upload_date_tag = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")

for i in Upload_date_tag:
    Upload_date.append(i.text)
```

```python
# Scraping Views

Views_tag = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")

for i in Views_tag:
    Views.append(i.text)
```

```python
# Creating dataframe for scraped data

Youtube = pd.DataFrame()
Youtube['Rank'] = Rank
Youtube['Name'] = Name
Youtube['Artist'] = Artist
Youtube['Upload date'] = Upload_date
Youtube['Views'] = Views
Youtube
```

```python
driver.close()
```

```python
# Scraping Skills they hire for

for i in driver.find_elements(By.XPATH,"//div[@class='hireSec highlightable']"):
    Skills_they_hire_for.append(i.text)
```

```python
# Scraping Location

from selenium.common.exceptions import NoSuchElementException

for i in range(1,len(Skills_they_hire_for)+1):
    if len(Location)!=len(Skills_they_hire_for):
        try:
            if (i%2)==0:
                Loc = driver.find_element(By.XPATH,f'/html/body/div[3]/div/div[2]/div/div[2]/div[{i}]/div[1]/div[1]/div[1]/p/span[2]/small')
                Location.append(Loc.text)
                
                Loc = driver.find_element(By.XPATH,f'/html/body/div[3]/div/div[2]/div/div[2]/div[{i}]/div[2]/div[1]/div[1]/p/span[2]/small')
                Location.append(Loc.text)
            
            else:
                Loc = driver.find_element(By.XPATH,f'/html/body/div[3]/div/div[2]/div/div[2]/div[{i}]/div[1]/div[1]/div[1]/p/span[2]/small')
                Location.append(Loc.text)
                
                Loc = driver.find_element(By.XPATH,f'/html/body/div[3]/div/div[2]/div/div[2]/div[{i}]/div[2]/div[1]/div[1]/p/span[2]/small')
                Location.append(Loc.text)
        
        except NoSuchElementException as e:
            print("Exception Raised:\n",e)
            Location.append("-")
            print("\nException handled successfully")
```

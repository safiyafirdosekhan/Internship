<b>2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
and make data frame.</b>

```python
page = requests.get('https://www.imdb.com/search/title/?count=100&groups=top_100&sort=user_rating')
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
name = []

for i in soup.find_all('h3',class_="lister-item-header"):
    name.append(i.text.replace("\n",""))

name = [i.split('.')[1] for i in name]
name = [i.split('(')[0] for i in name]    
```

```python
rating = []

for i in soup.find_all('div', class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.replace("\n",""))
```

```python
year_of_release = []

for i in soup.find_all('span', class_="lister-item-year text-muted unbold"):
     year_of_release.append(i.text)
```

```python
IMDB_Top_rated_100_movies = pd.DataFrame({'Name':name,'Rating':rating,'Year_of_Release':year_of_release,})
IMDB_Top_rated_100_movies[:100]
```

<b>5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:        
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.            
b) Top 10 ODI Batsmen along with the records of their team and rating.        
c) Top 10 ODI bowlers along with the records of their team and rating.</b>

<b>a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.</b>

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
team=[]
matches=[]
points=[]
rating=[]


# Take 1st Team, Matches, Points, Rating from Banner (table content)

for i in soup.find_all('span',class_="u-hide-phablet"):
        team.append(i.text)
    
for i in soup.find_all('td',class_="rankings-block__banner--matches"):
        matches.append(i.text)
    
    
for i in soup.find_all('td',class_="rankings-block__banner--points"):
        points.append(i.text)
    
for i in soup.find_all('td',class_="rankings-block__banner--rating u-text-right"):
        rating.append(i.text.strip())
    
# Take remaining Teams, Matches, Points and Rating (2-10)

table=[]
    
for i in soup.find_all('tr',class_="table-body"):
    table.append(i.text.split())

for i in table:
    matches.append(i[-3])
    points.append(i[-2])
    rating.append(i[-1])
```

```python
Top_10_ODI_Teams = pd.DataFrame({'Team':team[:10],'Matches':matches[:10],'Points':points[:10],'Rating':rating[:10]})
   
Top_10_ODI_Teams
```

<b>9) Write a python program to scrape mentioned details from dineout.co.in :        
i) Restaurant name       
ii) Cuisine        
iii) Location      
iv) Ratings       
v) Image URL</b>

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
restaurant_name = []

for i in soup.find_all('div', class_="restnt-info cursor"):
    restaurant_name.append(i.text)
```

```python
cuisine = []

for i in soup.find_all('span',class_="double-line-ellipsis" ):
    cuisine.append(i.text.split('|')[1])
```

```python
location = []

for i in soup.find_all('div',class_="restnt-loc ellipsis"  ):
    location.append(i.text)
```

```python
ratings = []

for i in soup.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.text)
```

```python
image_url = []

for i in soup.find_all('img',class_="no-img" ):
    image_url.append(i['data-src'])   
```

```python
Dineout = pd.DataFrame({'Restaurant Name':restaurant_name,'Cuisine':cuisine,'Location':location,'Ratings':ratings,'Image URL':image_url})
Dineout
```

<b>10) Write a python program to scrape the details of top publications from Google Scholar from
https://scholar.google.com/citations?view_op=top_venues&hl=en        
i) Rank       
ii) Publication      
iii) h5-index          
iv) h5-median</b>

```python
page = requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
rank = []

for i in soup.find_all("td",class_="gsc_mvt_p"):
    rank.append(i.text)
```

```python
publication = []

for i in soup("td",class_="gsc_mvt_t"): 
    publication.append(i.text)
```

```python
index = []

for i in soup.find_all("td",class_="gsc_mvt_n"):
    index.append(i.text)
    
h5_index = index[0::2] 

h5_median = index[1::2]
```

```python
Details_of_top_publications = pd.DataFrame({"Rank":rank,"Publication":publication,"h5-index":h5_index,"h5-median":h5_median})
Details_of_top_publications
```

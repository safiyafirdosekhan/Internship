<b>2. In the above question, now scrape the following details of each product listed in first 3 pages of your
search results and save it in a data frame and csv. In case if any product has less than 3 pages in search
results then scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.</b>

```python
# Collecting all the Product URLs

Product_URL = []

for i in range(0,3):
    Product_URL_tag = driver.find_elements(By.XPATH, "//a[@class='a-link-normal s-no-outline']")
    for i in Product_URL_tag:
        Product_URL.append(i.get_attribute('href'))
        
        
    # Clicking on the Next button 
    
    next_button = driver.find_element(By.XPATH, "//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']") 
    next_button.click()
    time.sleep(5)
```

```python
# Creating empty lists

Brand_Name= []
Name_of_the_Product = []
Price = []
Return_Exchange = []
Expected_Delivery = []
Availability = []
```

```python
# Start with for loop
for i in Product_URL:
    driver.get(i)
    
    
    
    # Scraping Brand Name and Name of the Product
    
    try:
        Name_of_the_Product_tag = driver.find_elements(By.XPATH, "//span[@id='productTitle']")
        for k in Name_of_the_Product_tag:
            P = k.text
            Name_of_the_Product.append(P)
            Name_of_the_Product = Name_of_the_Product[0:48]
    except NoSuchElementException as e:
        Name_of_the_Product.append("-")

    

    try:
        for k in range(len(Name_of_the_Product)):
            for j in Name_of_the_Product:
                brand = j.split(" ")
                Brand_Name.append(brand[0])
                Brand_Name = Brand_Name[0:48]
    except NoSuchElementException as e:
        Brand_Name.append("-")

    
    
    
    #Scraping Price
    try:
        Price_tag = driver.find_element(By.XPATH, "//span[@class='a-price-whole']")
        Price.append(Price_tag.text)
        Price = Price[0:48]
    except NoSuchElementException as e:
        Price.append("-")
        
    
        
    #Scraping Return/Exchange
    try:
        Return_Exchange_tag = driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[21]/div[2]/div/div/div/div[2]/span/div[2]/a")
        Return_Exchange.append(Return_Exchange_tag.text)
        Return_Exchange = Return_Exchange[0:48]
    except NoSuchElementException as e:
        Return_Exchange.append("-")
        
     
    
    #Scraping Expected_Delivery
    
    Expected_Delivery_tag = driver.find_element(By.XPATH,"//span[@class='a-text-bold']")
    try:
        Expected_Delivery.append(Expected_Delivery_tag.text)
        Expected_Delivery = Expected_Delivery[0:48]
    
    except NoSuchElementException as e:
        Expected_Delivery.append("-")
            
    
    
    
    #Scraping Availability
    try:
        Availability_tag = driver.find_element(By.XPATH, "//div[@id='availability']/span")
        Availability.append(Availability_tag.text)
        Availability = Availability[0:48]         
    except NoSuchElementException as e:
        Availability.append("-")
```

```python
len(Brand_Name), len(Name_of_the_Product), len(Price), len(Return_Exchange), len(Expected_Delivery), len(Availability[0:48]), len(Product_URL[0:48])
```

```python
# Data Framing

Laptop = {'Brand_Name' : Brand_Name[0:48], 'Name_of_the_Product':Name_of_the_Product[0:48], 'Price':Price[0:48], 'Return_Exchange':Return_Exchange[0:48], 'Expected_Delivery':Expected_Delivery[0:48], 'Availability':Availability[0:48], 'Product_URL':Product_URL[0:48]}

Laptopdf  = pd.DataFrame(Laptop)
Laptopdf
```

```python
# Creating empty list

funding={}
funding["Date"]=[]
funding["Startup Name"]=[]
funding["Industry or Vertical"]=[]
funding["Sub Vertical"]=[]
funding["Location"]=[]
funding["Investor"]=[]
funding["Investment Type"]=[]
funding["Amount(in USD)"]=[]
```

```python
# For second quarter (i.e., January 21 - March 21, the range numbers are between 54-57. 
# Let's take it and scrape the data.

for i in range(54,57):
    driver.find_element(By.XPATH,'//div[@id="tablepress-{}_wrapper"]/div/label/select/option[4]'.format(i)).click()
```

```python
for i in range(54,57):
    driver.find_element(By.XPATH,'//div[@id="tablepress-{}_wrapper"]/div/label/select/option[4]'.format(i)).click()

    
    
    # Scraping Date
    date_tags = driver.find_elements(By.XPATH,'//table[@id="tablepress-{}"]/tbody/tr/td[2]'.format(i))
    for date in date_tags:
        funding['Date'].append(date.text)

    
    
    # Scraping Startup Name
    name_tags = driver.find_elements(By.XPATH,'//table[@id="tablepress-{}"]/tbody/tr/td[3]'.format(i))
    for name in name_tags:
        funding['Startup Name'].append(name.text)
    
    
    
    # Scraping Industry Or Vertical
    ind_tags = driver.find_elements(By.XPATH,'//table[@id="tablepress-{}"]/tbody/tr/td[4]'.format(i))
    for n in ind_tags:
        funding['Industry or Vertical'].append(n.text)
    
    
    
    # Scraping Sub-Vertical
    sv_tags = driver.find_elements(By.XPATH,'//table[@id="tablepress-{}"]/tbody/tr/td[5]'.format(i))
    for sv in sv_tags:
        funding['Sub Vertical'].append(sv.text)

    
    
    # Scraping Location
    loc_tags = driver.find_elements(By.XPATH,'//table[@id="tablepress-{}"]/tbody/tr/td[6]'.format(i))
    for loc in loc_tags:
        funding['Location'].append(loc.text)
    
    
    
    # Scraping Investor
    inv_tags = driver.find_elements(By.XPATH,'//table[@id="tablepress-{}"]/tbody/tr/td[7]'.format(i))
    for inv in inv_tags:
        funding['Investor'].append(inv.text)
        
    
    
    # Scraping Investment Type
    invt_tags = driver.find_elements(By.XPATH,'//table[@id="tablepress-{}"]/tbody/tr/td[8]'.format(i))
    for invt in invt_tags:
        funding['Investment Type'].append(invt.text)
    
    
    
    # Scraping Amount
    amt_tags = driver.find_elements(By.XPATH,'//table[@id="tablepress-{}"]/tbody/tr/td[9]'.format(i))
    for amt in amt_tags:
        funding['Amount(in USD)'].append(amt.text)
```

```python
# Create dataframe

df=pd.DataFrame(funding)
df
```

<b> 10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.</b>

```python
# Connect to the webdriver

driver = webdriver.Chrome(r"C:\chromedriver.exe")
```

```python
# Get webpage

driver.get("https://www.hostelworld.com/")
driver.maximize_window()
```

```python
# Finding search menu

search = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[1]/div/div/div[4]/div/div[2]/div/div[1]/div/div/div/input')
search.send_keys("London")
```

```python
# Choosing first option

first_option = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[1]/div/div/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div').click()
```

```python
# Clicking on Let's go button

lets_go_button = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div[2]/div[1]/div/div/div[4]/div/div[2]/div/div[5]/button').click()
```

```python
# Creating Empty Lists

Hostel_Name = []
Distance = []
Ratings = []
Total_Reviews = []
Overall_Reviews = []
Privates = []
Dorms = []
Facilities = []
Property_Description = []
```

```python
# Scraping Names 
Hostel_Name_tag = driver.find_elements(By.XPATH,"//h2[@class='title title-6']")
for i in Hostel_Name_tag:
    Hostel_Name.append(i.text)
```

```python
# Scrapping Distance from city centre
try:
    Distance_tag = driver.find_elements(By.XPATH,'//span[@class="description"]')
    for i in Distance_tag:
        Distance.append(i.text)

except NoSuchElementException:
    Distance.append("-")
```

```python
# Scrapping Ratings
try:
    Ratings_tag = driver.find_elements(By.XPATH,'//div[@class="rating rating-summary-container big"]')
    for i in Ratings_tag:
        Ratings.append(i.text[:][0:3])

except NoSuchElementException:
    Ratings.append("-")
```

```python
# Scrapping Total Reviews
try:
    Total_Reviews_tag = driver.find_elements(By.XPATH,'//div[@class="reviews"]')
    for i in Total_Reviews_tag:
        Total_Reviews.append(i.text)

except NoSuchElementException:
    Total_Reviews.append("-")
```

```python
# Scrapping Overall Reviews
try:
    Overall_Reviews_tag = driver.find_elements(By.XPATH,'//div[@class="keyword"]')
    for i in Overall_Reviews_tag:
        Overall_Reviews.append(i.text)

except NoSuchElementException:
    Overall_Reviews.append("-")
```

```python
# Scrapping Privates from price
Privates_tag = driver.find_elements(By.XPATH,'//div[@class="price-col"]')[0::2]
for i in Privates_tag:
    Privates.append(i.text.replace('\n',': ').replace('No Privates Available','NaN'))
```

```python
# Scrapping Dorms from price
Dorms_tag = driver.find_elements(By.XPATH,'//div[@class="price-col"]')[1::2]
for i in Dorms_tag:
    Dorms.append(i.text.replace('\n',': ').replace('No Dorms Available','NaN'))
```

```python
# Scrapping Facilities
try:
    Facilities_tag = driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
    for i in Facilities_tag:
        Facilities.append(i.text.replace('\n',', '))

except NoSuchElementException:
    Facilities.append("-")
```

```python
# Scrapping Property Description.
try:
    Property_Description_tag = driver.find_elements(By.XPATH,'//div[@class="rating-factors prop-card-tablet rating-factors small"]')
    for i in Property_Description_tag:
        Property_Description.append(i.text.replace('\n',', '))

except NoSuchElementException:
    Property_Description.append("-")
```

```python
print(len(Hostel_Name), len(Distance), len(Ratings), len(Total_Reviews), len(Overall_Reviews), len(Privates), len(Dorms), len(Facilities), len(Property_Description))
```
